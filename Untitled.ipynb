```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('healthcare-dataset-stroke-data.csv')
df.head()
```

```python
df.shape
```

```python
#missing value check
missing= df.isna().sum()
missing
```

```python
values_to_fill={'bmi':'no bmi'}
df.fillna(value=values_to_fill,inplace=True)
```

```python
number_of_missig_values= df.isna().sum()
number_of_missig_values
```

```python
room_type_counts = df['smoking_status'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('x')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['avg_glucose_level'], alpha=0.5, color='black')
plt.xlabel('age')
plt.ylabel('avg_glucose_level')
plt.title('age vs hypertension')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
```

```python
columns = ['age', 'hypertension', 'heart_disease', 'avg_glucose_level','stroke']
required_data= df[columns]
new_column_names = ['Age', 'Hypertension', 'Heart disease', 'Average glucose level','Stroke']
required_data.columns = new_column_names

correlation_matrix = required_data.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap')
plt.show()
```
